```python
import os
import numpy as np
from datasets import Dataset
from transformers import AutoTokenizer
from dotenv import load_dotenv
load_dotenv()

token = os.getenv("HF_TOKEN")

token_id = "meta-llama/Meta-Llama-3-8B"
tokenizer = AutoTokenizer.from_pretrained(token_id, token=token)
```

```python
# Sample TTR calcualtion

sample_text = "Artificial intelligence is transforming the world."

print("Original text:")
print(sample_text)

print('\nTokenized text')
tokens = tokenizer.tokenize(sample_text)
print(tokens)

token_len = len(tokens)
print(token_len)

types = len(set(tokens))
print(types)

ttr = (types / token_len) * 100
print(f"TTR = {ttr}%")
```
